```python
emulator = Emulator(number_of_point=500, sampling_methode='lhs', kernel_type='Matern')
```

```python
samples = emulator.sampling().drop(index=3)
```

```python
test_emul = Emulator(number_of_point=31, sampling_methode='random', kernel_type='Matern')
test_set = test_emul.sampling()
```

```python
emulator.train(train_df = samples)
```

```python
emulator.predict([[0.4,0.23,0.5,0.8]])
```

```python
index = 11
y_pred = emulator.predict([test_set[['alpha','beta','delta','gamma']].values[index]]).reshape((21,1))
y_true = test_set.prey.values[index]
print(test_set[['alpha','beta','delta','gamma']].values[index])
```

```python
plt.plot(y_pred, label ='pred')
plt.plot(y_true, label='true')
plt.legend()
```
