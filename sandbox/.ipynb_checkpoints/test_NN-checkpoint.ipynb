# Test with $ y_{scaled}=\frac{y-y_{min}}{y_{max}-y_{min}} $

```python
inputA = Input(shape=(31,6))
x = Dense((64), activation="relu")(inputA)
x = LSTM(64, return_sequences=True,input_shape=(31,6))(x)
x = Dense(2)(x)
x = Reshape((2,31))(x)

model_t_series_norm = Model(inputs=[inputA], outputs=x)
model_t_series_norm.compile(loss='MSE', optimizer='adam')
model_t_series_norm.summary()
```

```python
inputB = Input(shape=(6,))
y = Dense(128,activation='relu')(inputB)
y = Dense(32,activation='relu')(y)
y = Dense(8,activation='relu')(y)
y = Dense(4)(y)
model_y_max = Model(inputs=[inputB], outputs=y)
model_y_max.compile(loss='mae', optimizer='adam')
model_y_max.summary()
```

```python
def reshape_input(x,time_size):
    x_n = np.repeat(x[:, :, np.newaxis], time_size , axis=2)
    return np.array([np.array(x_n[i].T) for i in range(len(x_n))]), x
```

```python
def reshape_output(y):
    y = np.array([np.array(([np.array(y[i][0]),np.array(y[i][1])])) for i in range (len(y))])
    y_max  = np.max(y,axis=2).reshape(len(y),2)
    y_min = np.min(y, axis= 2).reshape(len(y),2)
    for i in range(len(y)):
        y[i][0] = (y[i][0] - y_min[i][0])/np.abs(y_max[i][0]-y_min[i][0])
        y[i][1] = (y[i][1] - y_min[i][1])/np.abs(y_max[i][1]-y_min[i][1])
    y_norm = y
    return y_norm, y_max ,y_min

def drop_outliers_max(x,y_max_min):
    df= pd.DataFrame(np.concatenate([x,y_max_min],axis=1), columns=['alpha','beta','delta','gamma','prey_0','predator_0','max_prey','max_predator','min_prey','min_predator'])
    df = df[(df['max_prey']< df['max_prey'].quantile(0.99))]
    df = df[(df['max_predator']< df['max_predator'].quantile(0.99))]
    return df[['alpha','beta','delta','gamma','prey_0', 'predator_0']].values, df[['max_prey','max_predator','min_prey','min_predator']].values
```

# Test with $y_{scaled} = \frac{y-y_0}{amplitude} $  
$ amplitude = y_{max}-y_{min}$

```python
inputA = Input(shape=(31,6))
x = Dense((128), activation="relu")(inputA)
x = LSTM(64, return_sequences=True,input_shape=(31,6))(x)
x = Dense(2)(x)
x = Reshape((2,31))(x)

model_t_series = Model(inputs=[inputA], outputs=x)
model_t_series.compile(loss='MSE', optimizer='adam')

model_t_series.summary()
```

```python
inputB = Input(shape=(6,))
y = Dense(128,activation='relu')(inputB)
y = Dense(32,activation='relu')(y)
y = Dense(8,activation='relu')(y)
y = Dense(2,activation='relu')(y)
model_alpha = Model(inputs=[inputB], outputs=y)
model_alpha.compile(loss='mae', optimizer='adam')
model_alpha.summary()
```

```python
def reshape_output_amplitude(y):
    y = np.array([np.array(([np.array(y[i][0]),np.array(y[i][1])])) for i in range (len(y))])
    y_max  = np.max(y,axis=2).reshape(len(y),2)
    y_min = np.min(y, axis= 2).reshape(len(y),2)
    amplitude = y_max - y_min
    
    for i in range(len(y)):
        y[i][0] = (y[i][0] - y[i][0][0])/amplitude[i][0]
        y[i][1] = (y[i][1] - y[i][1][0])/amplitude[i][1]
    y_norm = y
    return y_norm, amplitude
```

```python
def drop_outliers_amplitude(x,amplitude):
    df= pd.DataFrame(np.concatenate([x,amplitude],axis=1), columns=['alpha','beta','delta','gamma','prey_0','predator_0','amplitude_prey','amplitude_predator'])
    df = df[(df['amplitude_prey']< df['amplitude_prey'].quantile(0.99))]
    df = df[(df['amplitude_predator']< df['amplitude_predator'].quantile(0.99))]
    return df[['alpha','beta','delta','gamma','prey_0', 'predator_0']].values, df[['amplitude_prey','amplitude_predator']].values
```
