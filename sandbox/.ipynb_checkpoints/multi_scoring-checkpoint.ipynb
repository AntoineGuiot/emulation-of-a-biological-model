```python
%load_ext autoreload

%autoreload 2
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import time
from emulator import Emulator
from model import Model
from vpop import Vpop
from voting_system import VotingSystem
```

```python
descriptors = ["alpha", "beta", "delta", "gamma", "DefaultComp.predator", "DefaultComp.prey"]

# create vpop testing
vpop_test = Vpop(f'vpop testing',descriptors=descriptors,number_of_point=149, min_=0.1, max_=1,distribution=True, sampling_method='random')
vpop_test.sample()
# create model
model = Model('model')

# create training vpop
vpop_train = Vpop(f'vpop train', descriptors=descriptors,
                  number_of_point=3000, min_=0.1, max_=1, distribution=True)
vpop_train.sample()

# run model on vpop
training_data_frame = model.run(vpop_train)
print('train sampling done')

testing_data_frame = model.run(vpop_test)
print('test sampling done')

# create Voting system
voting_system = VotingSystem('', 5, training_data_frame, x_parameters=descriptors, t_series_names=['prey', 'predator'])
```

```python
voting_system.start_training()
```

```python
Res = voting_system.predict(np.array([[0.4,0.3,0.12,0.34,1.3,1.2],[0.4,0.3,0.12,0.34,1.3,14]]))
```
